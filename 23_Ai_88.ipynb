```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import VarianceThreshold
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
v_thershold = VarianceThreshold(threshold=0)
v_thershold.fit(df[['math score','reading score','writing score']])
```

```python
for i in df:
  if df[i].dtype == 'int64':
    df=df.fillna(df[i].mean())

  else:
    df=df.fillna(df[i].mode())

print(df)
```

```python
for i in df:
  if df[i].dtype == 'int64':
    df=df.fillna(df[i].mean()/df[i].std())
df
```

```python
plt.hist(df['math score'])
plt.show()
```
